```python
# •	Convert categorical features to numeric using pd.get_dummies.
# •	Create a bar plot that shows feature importance.
# •	Example code to get feature importance:

importances = clf.feature_importances_

feature_importances = pd.DataFrame({'Feature': df_filter.columns, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

seaborn.barplot(x = 'Feature',
            y = 'Importance',
            data = feature_importances).set(title='Feature Importance')

# Show the plot
plt.xticks(rotation = 90)
plt.show()
```
