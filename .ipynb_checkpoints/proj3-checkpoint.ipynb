```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

data = pd.read_csv('troop_movements.csv')

homeworlds = data['homeworld']
unit_types = data['unit_type']
affiliations = data['empire_or_resistance']

le_homeworld = LabelEncoder()
le_unit_type = LabelEncoder()
le_affiliation = LabelEncoder()

homeworlds_encoded = le_homeworld.fit_transform(homeworlds)
unit_types_encoded = le_unit_type.fit_transform(unit_types)
affiliations_encoded = le_affiliation.fit_transform(affiliations)

features = list(zip(homeworlds_encoded, unit_types_encoded))

X_train, X_test, y_train, y_test = train_test_split(features, affiliations_encoded, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

predicted_affiliations = le_affiliation.inverse_transform(y_pred)

accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)
```
